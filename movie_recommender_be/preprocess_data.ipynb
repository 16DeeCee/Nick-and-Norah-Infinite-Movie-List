```python
import pandas as pd
```

```python
data_df = pd.read_csv("raw_data/TMDB_movie_dataset_v11.csv")
```

```python
data_df.dropna(subset=[
    "title",
    "overview",
    "genres",
    "keywords", 
    "poster_path",
    "release_date",
    "imdb_id",
    "poster_path", 
    "keywords",
    "popularity",
    "status",
    "vote_average",
    "vote_count",
], 
ignore_index=True, 
inplace=True
)
```

```python
data_df = data_df.loc[data_df["adult"] == False, :]
data_df = data_df.loc[data_df["status"] == "Released", :]
data_df = data_df.loc[data_df["vote_count"] > 0]
```

```python
data_df.head(10)
```

```python
# data_df.loc[:, "genres"] = data_df["genres"].map(lambda x: x.split(", "))
```

```python
data_df["poster_path"] = "https://image.tmdb.org/t/p/w500" + data_df["poster_path"]
```

```python
data_df["release_date"] = data_df["release_date"].apply(lambda x: int(x.split("-")[0]))
```

```python
data_df.rename(columns={"release_date": "release_year", "vote_average": "rating"}, inplace=True)
```

```python
data_df = data_df.reset_index(drop=True)
```

```python
data_df.drop(columns=[
    "status", 
    "revenue",
    "adult",
    "backdrop_path", 
    "budget", 
    "homepage", 
    "original_language", 
    "original_title", 
    "tagline",
    "production_companies",
    "production_countries",
    "spoken_languages",
], 
axis=1,
inplace=True
)
```

```python
mean_votes = data_df["vote_count"].mean()
mean_rating = data_df["rating"].mean()

print(f"Average number of ratings for a given movie: {mean_votes:.2f}")
print(f"Average rating for a given movie: {mean_rating:.2f}")
```

```python
def bayesian_average(rating, num_votes):
    bayes_avg = (mean_votes * mean_rating + (rating * num_votes)) / (mean_votes + num_votes)

    return round(bayes_avg, 2)
```

```python
data_df["bayesian_rating"] = data_df.apply(lambda x: bayesian_average(x["rating"], x["vote_count"]), axis=1)
```

```python
data_df.isna().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

data_df["normalized_bayes_rating"] = scaler.fit_transform(data_df[["bayesian_rating"]])
data_df["normalized_popularity"] = scaler.fit_transform(data_df[["popularity"]])
```

```python
data_df.to_csv("preprocessed_data/clean_movie_dataset.csv", index=False)
```

```python
data_df["full_details"] = "Title: " + data_df["title"].astype(str) + " | Overview: " + data_df["overview"].astype(str)  + " | Genres: " + data_df["genres"].astype(str) + " | Keywords: " + data_df["keywords"].astype(str)
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
embeddings = model.encode(data_df["full_details"], convert_to_numpy=True)
```

```python
import faiss
import numpy as np

dimension = embeddings.shape[1]
faiss.normalize_L2(embeddings)
index = faiss.IndexFlatIP(dimension) # Inner Product
index.add(embeddings.astype(np.float32))
```

```python
faiss.write_index(index, "faiss_index.idx")
```
